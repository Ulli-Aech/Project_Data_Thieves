```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
url = 'http://wdi.worldbank.org/table/3.4'
html = requests.get(url).content
soup = BeautifulSoup(html, "html.parser")
soup
```

```python
table = soup.find_all('div',{'class':'scrollable'})[0]
table
```

```python

rows = table.find_all('tr')
rows = [row.text.strip().split("\n") for row in rows]
rows
```

```python

df = pd.DataFrame(rows)

df

#[[:alpha:]]+/g

#df["countries"]=df[0].str.extract('([A-Za-z]+[A-Za-z])')

df["countries"]=df[0].str.extract('(\D+)')

df

#df["numeric"]=df[0].str.extract('(\d+\s+\.\,)')

df["numeric"]=df[0].str.replace('([A-Za-z])',' ')

df.columns

df
```

```python
new=df['numeric'].str.split(expand=True)
#data.assign(new)
#new+df['countries']
data=pd.concat([df,new],axis=1)
data
```

```python
#creating a copy 

forest=data.copy()
forest
```

```python
#check zeros

none_columns=forest.isnull().sum()
none_columns[none_columns > 0]
```

```python
#dropping columns with a lot of zeros

drop_cols = list(none_columns[none_columns > 190].index)
forest = forest.drop(drop_cols, axis=1)

forest
```

```python
#check for duplicates

before=len(forest)

forest.drop_duplicates(inplace=True)

after = len(forest)
print('Number of duplicate records dropped: ', str(before - after))
```
